# 起步
你将运行自己的第一个程序——hello_world.py。如果你想本机访问，首先需要检查自己的计算机是否安装了Python;如果没有安装，你需要安装它。

## 搭建编程环境
在不同的操作系统中，Python存在细微的差别，因此有几点你需要牢记在心。这里将介绍大家使用的两个主要的Python版本，并简要介绍Python的安装步骤。

### Python 2和Python 3
当前，有两个不同的Python版本:Python 2和较新的Python 3。每种编程语言都会随着新概念和新技术的推出而不断发展，Python的开发者也一直致力于丰富和强化其功能。大多数修改都是逐步进行的，你几乎意识不到，但如果你的系统安装的是Python 3，那么有些使用Python 2编写的代码可能无法正确地运行。在本书中，我将指出Python 2和Python 3的重大差别，这样无论你安装的是哪个Python版本，都能够按书中的说明去做。  
如果你的系统安装了这两个版本，请使用Python 3;如果没有安装Python，请安装Python 3;如果只安装了Python 2，也可直接使用它来编写代码，但还是尽快升级到Python 3为好，因为这样你就能使用最新的Python版本了。

## 在不同操作系统中搭建Python编程环境
Python是一种跨平台的编程语言，这意味着它能够运行在所有主要的操作系统中。在所有安装了Python的现代计算机上，都能够运行你编写的任何Python程序。然而，在不同的操 作系统中，安装Python的方法存在细微的差别。     
在这一节中，你将学习如何在自己的系统中安装Python和运行Hello World程序。

#### 在OS X系统中安装Python 3
如果你的系统只安装了Python 2，或者已安装的Python 3版本较旧，可使用一个名为Homebrew的包来安装最新的Python 3版本。 
1. 安装Homebrew  
Homebrew依赖于Apple包Xcode，因此请打开一个终端窗口并执行如下命令:  
```
$ xcode-select --install
```
在不断出现的确认对话框中都单击OK按钮(根据网络连接的速度，这可能会花一些时间)。接下来安装Homebrew:
```
$ ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)"
```
这个命令可在Homebrew网站(http://brew.sh/)的首页找到。在curl -fsSL和URL之间，务必包含一个空格。
2. 安装Python3  
为安装最新的Python 3版本，请执行如下命令:
```
$ brew install python3
```
下面来检查使用这个命令安装的是哪个版本:
```
$ python3 --version
Python 3.8.2
$
```

如果你的windows系统是64位，选择[Windows x86-64 executable installer](https://www.python.org/ftp/python/3.8.3/python-3.8.3-amd64.exe)；如果你的windows系统是32位，选择[Windows x86 executable installer](https://www.python.org/ftp/python/3.8.3/python-3.8.3.exe)  
下载完Python的安装文件，双击安装，在安装界面确保勾选了<input type='checkbox' checked='true'/>`Add Python 3.8.3 to PATH`。

## 编写、运行第一个Python程序
编写、运行第一个hello world级的Python程序，有以下3种方式：

### 在终端交互式界面编写、运行程序
无论是linux、mac os还是windows，你都可以打开一个终端命令行窗口，输入`python`命令进入python的交互式界面

在“>>>”提示符后面交互式输入程序代码，由于python是解释型编程语言，每次输入完代码，敲回车键，即可看到运行结果。
```
>>> print('hello world')
hello world
>>> 
```

### 在IDE（集成开发环境）中编写、运行程序
Python的集成开发工具很多，如自带的IDLE、PyCharm、VsCode等等。

#### IDLE
大部分情况下，当你安装完Python，都会带有一个简易的IDE工具，名字叫“IDLE”，是一款极轻量级的开发工具，对于简单的学习来说，足够用了。

找到并运行它，你将会看到如下的界面，默认已打开一个shell命令行交互窗口，用法如同上面的

#### PyCharm
PyCharm是捷克[JetBrains](https://www.jetbrains.com)公司出品的一款功能极为强大的Python开发工具。进入[下载](https://www.jetbrains.com/pycharm/download)页面

你会看到PyCharm支持Windows、Mac和Linux操作系统，PyCharm版本分为收费的`Professional`专业版和免费的`Community`社区版，对于一般的学习开发而言，`Community`社区版已经足够好用了。至于如何下载、安装、使用，因为非常简单，这里不再赘言。
